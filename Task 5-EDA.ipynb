Now let us analyse the profit with differnt categories. 
We will use Catplot for the analysis


```python
#Profit vs Shipmode
sns.catplot(x ="Ship Mode", y="Profit", data = analysis, color='red', height=3, aspect=3);
```

Now we conclude that Standard Class Shipmode is more profitable.

```python
#profit vs Segment 
sns.catplot(x ="Segment", y="Profit", data = analysis, color='blue', height=3, aspect=3);
```

We can conclude that we are earning more profit from Consumer Segment

```python
#Profit vs Region
sns.catplot(x ="Region", y="Profit", data = analysis, color='Green', height=3, aspect=3);
```

The eastern region is more profitable than any other region

```python
#Profit vs Category
sns.catplot(x ="Category", y="Profit", data = analysis, color='Yellow', height=3, aspect=3);
```

We are earning more profit from Technology category
